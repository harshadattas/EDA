```python
final_df = pd.merge(df,df_country,on = 'Country Code',how = 'left')
final_df.head(2)
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
```

Observations : 
1. It is evident that zomato's transaction are maximum in India
2. USA and UK take the 2nd and 3rd spots respectively

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings.head()
```

Observations : 
1. When rating is between 4.5 - 4.9 ----> Excellent
2. When rating is between 4.0 - 4.4 ----> Very good
3. When rating is between 3.5 - 3.9 ----> Good
4. When rating is between 3.0 - 3.4 ----> Average
5. When rating is between 2.5 - 2.9 ----> Average
6. When rating is between 1.8 - 2.4 ----> Poor

```python
plt.figure(figsize=(14,8))
sns.barplot(x="Aggregate rating", y= "Rating Count", data=ratings)
```

```python
sns.countplot(x = 'Rating color', data = ratings)
```

```python
# Countries that have given 0 rating 

final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

```python
# Find out which currency is used by which country

final_df[['Currency', 'Country']].groupby(['Currency', 'Country']).size().reset_index()
```

```python
# Which countries do have online delivery option

final_df[final_df['Has Online delivery'] == 'Yes'].groupby('Country').size().reset_index()
```

```python
city_values = final_df.City.value_counts().values
city_labels = final_df.City.value_counts().index
```

```python
# pie chart for city distribution

plt.pie(city_values[:5], labels = city_labels[:5], autopct='%1.2f%%')
```
